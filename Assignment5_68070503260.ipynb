```python
#1Define a class named Car.
#2 add an initializer(_init_) method with three attributes:
   # model (String)
   # rent_per_day (float)
   # available (boolean, defaut value = True)
#3 Add a method rent (days) that:
   # if available is True, set it to false, calculate total cost(rent_per_day = days),
   # and print "Car rented. Total cost = " with the cost.
   # Otherwise, print "Car not available"
#4. Add a method return_car() that: (4 marks)
   #o Sets available to True
   #o Prints "Car returned"
#5. Add a method display() that prints the car model, rent per day, and availability status.
#6. In the main program: (9 marks)
   #o Create an object for Car with a model and rent per day. (2 marks)
   #o Call the rent method for 5 days. (2 marks)
   #o Call the display method. (2 marks)
   #o Call the return_car method and display the car info again. (3 marks)

class Car:
    def __init__(self, model, rent_per_day, available):
        self.model = model
        self.rent_per_day = rent_per_day
        self.available = True
    
    def rent(self, days):
        if self.available:
            total_cost = self.rent_per_day * days
            self.available = False
            print("Car Rented. Total cost=", total_cost)
        else: 
            print("Car is not available to rent")

    def return_car(self):
        self.available = True
        print("Car returned")

    def display(self):
        print("Model =", self.model)
        print("Rent =", self.rent_per_day)
        print("Available =", self.available)

my_car = Car("BMW", 3000, True)
my_car.rent(5)
my_car.display()
my_car.return_car()
my_car.display()
    



    

        
        
```

```python
# ## Inheritance
# #Step 1 Parent Class Person
# 1. Define a class "Person",
# 2. Add_init_(self, name, age) and store those values ​​on the object.
# 3. Add a method display_info(self) that prints: Name: <name>, Age: <age>
# Step 2: Child Class Student
# 1. Define a class "Student" that inherits from "Person".
# 2. Add an init function with super() to inherit the attributes from the parent class.
# 3. Add a new attribute "marks" and set it to an empty dictionary {}:
# e.g., {"Math": 85}
# 4. Implement add_mark(self, subject, score) function inside the Student class:
# Add or update the score in the marks dict.
# If the subject does not exist in the mark dictionary, add a new subject and score to the dictionary.
# Otherwise, update the score of the existing subject
# 90" print a confirmation message like "Added Science = 85" or "Updated Math"
# 5. Implement display_marks(self) function.
# 6. Implement average(self):
# If no marks, return 0
# otherwise, return the average (sum of scores / number of subjects)
# Step 3: Main Program
# 1. Create one student with name and age.
# 2. Call display_info()
# 3. Add at least two subject marks
# 4. Call display_marks()
# 5. Print the student's average score

class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def display_info(self):
        print("Name = ", self.name)
        print("Age = ", self.age)
    
class student(person):
    def __init__(self, name, age,):
        super().__init__(name, age)
        self.marks = {}
        
      
    def add_marks(self, subject, score):
        self.marks[subject] = score
        if subject in self.marks:
            print(f"Updated {subject} = {score}")
        else:
            print(f"Added {subject} = {score}")
        
    def display_marks(self):
        if not self.marks:
            print("No marks here.")
            return
        print ("Marks:")
        for subject, score in self.marks.items():
            print(f" - {subject}: {score}")
       
    def average(self):
        if not self.marks:
            return 0
        average = sum(self.marks.values()) / len(self.marks)
        return average

if __name__ == "__main__":
    student1 = student("Jack", 17)
    student1.display_info()

    student1.add_marks("Math", 80)
    student1.add_marks("Science", 95)

    student1.display_marks()
    print("The student average score is: ", student1.average())

```
